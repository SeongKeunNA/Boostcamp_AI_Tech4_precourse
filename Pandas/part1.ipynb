index를 문자로 바꿀 수 있는 것이 리스트와 가장 큰 차이점, numpy.ndarray의 서브클래스

```python
list_name = ['a','b','c','d','e']
example_obj = Series(data=  list_data, index = list_name)
example_obj
```
